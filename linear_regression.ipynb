```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('polydata.csv')
# def loss_function(m , b, data):
#     total_error = 0
#     for i in range(len(data)):
#         x = data.iloc[i].age
#         y = data.iloc[i].height
#         total_error += (y - (m*x+b)) ** 2
#     total_error/ float(len(data))

def gradient_descent(m_now, b_now, data, L):
    m_gradient = 0
    b_gradient = 0
    n = len(data)
    for i in range(n):
        x = data.iloc[i].age
        y = data.iloc[i]["height(cm)"]

        m_gradient+= -(2/n)*x*(y - (m_now*x+b_now))
        b_gradient+= -(2/n)*(y - (m_now*x+b_now))

    m = m_now - m_gradient*L
    b = b_now - b_gradient*L
    return m, b

m = 0
b = 0
L = 0.015
epochs = 1000

for i in range (epochs):
    if i % 50 == 0:
        print(f"Epoch:{i}")
    m, b = gradient_descent(m,b,df,L)


print(m, b)
plt.scatter(df.age, df["height(cm)"], color = "black")
plt.plot(list(range(0, 6)), [m*x+b for x in range(0, 6)], color ="red")
plt.show()
```
